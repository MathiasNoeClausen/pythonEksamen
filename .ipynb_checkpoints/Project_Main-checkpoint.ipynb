På vivino kan brugere sammen med deres review indtaste hvor vinen ligger på 3 forskellige skalaer: 

Let ---> Kraftig

Tør ---> Sød

Blød ---> Sur

Vi ville undersøge forholdende mellem disse notes og andre columns såsom lande og vintyper

```python
%pylab inline 
from py_files.Wine_Notes_Med_3d import get_appended_df, create_3d_plot, create_predicted_3d_plot, plot_mean_notes_for_countries, plot_mean_notes_for_winetypes, get_correlation_between_country_and_notes
```

```python
appended_df = get_appended_df()
```

```python
create_3d_plot(appended_df)
```

Hvor rød er rødvine, grøn er hvidvine, blå er rosé vine og cyan er dessert vine. Som forventet har desser vine en tendens at være mere sød, mens rosé vine sjældent har en score på højere end 50 i kraftighed. Rødvine og hvidvine er mere spredt men gennemsnitligt er rødvin mere kraftig og har flere outliers med meget lavt syre niveau. 
 
Modellen ligger også i '3d presentation notes' hvor man burde kunne bevæge rundt i den.

```python
create_predicted_3d_plot(appended_df)
```

## Best wine by year


```python
from py_files.BestWineByYear import max_rating_df, plot_max_df, max_occurences
```

```python
#Renser i datasettet for kun at få fat i den data, vi skal bruge. 
#Vin skal være fra før 2022 og have mere end 0 reviews for at indgå i vores dataframe.
#Derudover fjerner vi Price kolonnen, da den ikke er relevant i forhold til det vi søger efter.
year_review_mask = (df1['Year'] < 2022) & (df1['Reviews'] > 0)
dfClean = df1[year_review_mask]


data_chart = dfClean.drop(columns=('Price'))
data_chart.head()
```

```python
#Test index på vin, for at sikre at de forskellige kolonner ikke er blandet sammen med max funktionen.
data_chart.loc[1924]
```

```python
#Dette dataframe viser hvilke vine, der har haft de højeste ratings fra 1982-2020.
maxdf = max_rating_df(data_chart)
maxdf
```

```python
#Dette er et bar-chart der tager maxdf, som parameter for at visualisere dataframet. 
#Derudover viser det hvor mange reviews de enkelte vine, og man kan derfor se en sammenhæng mellem ratings og reviews. 
maxdfplot = plot_max_df(maxdf)
maxdfplot
```

```python
#Viser hvilke lande der flest gange har haft den/en af de bedste vin for et bestemt år. 
max_occurences(maxdf)
```
